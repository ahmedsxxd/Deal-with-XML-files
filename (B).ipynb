```python
import glob
import pandas as pd
import xml.etree.ElementTree as ET 
from datetime import datetime
```

```python
def extract_from_xml(file_to_process):
    dataframe = pd.DataFrame(columns=["question_id", "media_identifier", "question_title", "start_time", "end_time"])
    tree = ET.parse(file_to_process)
    root = tree.getroot()

    for question in root.find('transcript').findall('question'):
        try:
            question_id = question.attrib['id']
        except:
            question_id = None
        try:
            media_identifier = question.attrib["media_identifier"]
        except:
            media_identifier = None
        try:
            question_title = question.find('question_title').text
        except:
            question_title = None
        start_time = question.find("p").find("s").attrib["start_time"]
        end_time = question.find("p").find("s").attrib["end_time"]

        print(f"id: {question_id}")
        print(f"media_identifier: {media_identifier}")
        print(f"titel: {question_title}") 
        print(f"start_time: {start_time}")
        dataframe = dataframe.append({"question_id":question_id, "media_identifier":media_identifier, "question_title":question_title, "start_time":start_time, "end_time":end_time}, ignore_index=True)
    return dataframe
```

```python
def extract():
    extracted_data = pd.DataFrame(columns=["question_id", "media_identifier", "question_title","start_time", "end_time"])
    #process all xml files
    for xmlfile in glob.glob("*.xml"):
        print(f"file name: {xmlfile}")
        extracted_data = (extract_from_xml(xmlfile))
        extracted_data.to_csv(f'output files/{xmlfile}.csv', index=False)
        print('___' * 30)
        
    return extracted_data
```

```python
df = extract()
```
