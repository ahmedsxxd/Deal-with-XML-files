## 1)	Copy the XML files to a new path titled “Updated”


```python
import glob
import pandas as pd
import xml.etree.ElementTree as ET 
import os
from shutil import copy as cp
```

```python
def copy_XML_files():
    try:
        os.mkdir("Updated")
    except:
        print("directory oredy created")
        
    for xmlfile in glob.glob("*.xml"):
        print(f"file name: {xmlfile}")
        cp(xmlfile, "Updated")
        print('___' * 20)  
```

```python
copy_XML_files()
```

## change from p_tag


```python
def change_p_tag(file_to_process):
    tree = ET.parse(file_to_process)
    root = tree.getroot()
    new_element = ET.Element("s")
    for paragraph in root.find('body').findall('p'):
        sub_element = ET.SubElement(new_element, 's')
        sub_element.text = paragraph.text
        paragraph.clear()
        paragraph.append(sub_element)
        ET.dump(paragraph)
        
```

```python
def change_p(): 
    for xmlfile in glob.glob("Updated/*.xml"):
        print(f"file name: {xmlfile}")
        change_p_tag(xmlfile)
        print('___' * 40)
        
```

```python
change_p()
```

## change from td_tag

```python
def change_td_tag(file_to_process):
    tree = ET.parse(file_to_process)
    root = tree.getroot()
    new_element = ET.Element("s")
    for table in root.find('body').findall('table'):
        for table_row in table.iter("tr"):
            for table_data in table_row.iter("td"):
                sub_element = ET.SubElement(new_element, 's')
                sub_element.text = table_data.text
                table_data.clear()
                table_data.append(sub_element)
                ET.dump(table_data)
```

```python
def change_td(): 
    for xmlfile in glob.glob("Updated/*.xml"):
        print(f"file name: {xmlfile}")
        change_td_tag(xmlfile)
        print('___' * 40)
        
```

```python
change_td()
```
