```python
#Import the required modules and functions
import glob
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
def extract_from_xml(file_to_process):
    dataframe = pd.DataFrame(columns=["id", "name", "seo_meta_description", "source_id", "developer_name"])
    tree = ET.parse(file_to_process)
    root = tree.getroot()
    try:
        _id = root.attrib['id']
    except:
        _id = None
    try:
        name = root.find("title").text
    except:
        name = None
    try:
        seo_meta_description = root.find("seo_meta_description").text
    except:
        seo_meta_description = None
    try:
        source_id = root.find("source_id").text
    except:
        source_id = None
    try:
        developer_name = root.find("developer_name").text
    except:
        developer_name = None
    print(f"id: {_id}")
    print(f"name: {name}")
    print(f"seo_meta_description: {seo_meta_description}")
    print(f"source_id: {source_id}")
    print(f"developer_name: {developer_name}")
    dataframe = dataframe.append({"id":_id, "name":name, "seo_meta_description":seo_meta_description, "source_id": source_id, "developer_name":developer_name}, ignore_index=True)
    return dataframe
```

```python
def extract():
    extracted_data = pd.DataFrame(columns=["id", "name", "seo_meta_description", "source_id", "developer_name"])
    #process all xml files
    for xmlfile in glob.glob("*.xml"):
        print(f"file name: {xmlfile}")
        extracted_data = extracted_data.append(extract_from_xml(xmlfile))
        print('___' * 20)
        
    return extracted_data
```

```python
df = extract()
```
